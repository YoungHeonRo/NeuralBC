```python
import pymongo
from pymongo import MongoClient

# MongoDB Connect
Connection = MongoClient('52.79.241.214', 27017)

# MongoDB Databases Create and use
Database = Connection['coiner']

# MongoDB Collection Create and use
Collection = Database['price_info']

# Insert Document
Collection.insert_one({'1':'hi', '2':'hello', '3':'nice', '4':25})
Collection.insert_one({'5':'hi', '6':'hello', '7':'nice', '8':25})
Collection.insert_one({'9':'hi', '10':'hello', '11':'nice', '12':25})
Collection.insert_one({'15':'hi', '25':'hello', '32':'nice', '32':25})
```

```python
# find Document
result = Collection.find()

# Print(result)
for i in result:
    print(i)
```

```python
# Conditional Search
Search = Collection.find({'empno':'10003'})

# Print(Search)
for j in Search:
    print(j)
```

```python
# Delete Collection
Collection.delete_one({'9':'hi'})
Collection.find_one({'1':'hi'})
```
